# Create Model

```python
x=data.drop('Class',axis=1)
y=data['Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
model1=LogisticRegression()
```

```python
def cal(model):
        model.fit(x_train,y_train)
        pre=model.predict(x_test)
        accuracy=accuracy_score(pre,y_test)
        recall=recall_score(pre,y_test)
        f1=f1_score(pre,y_test)
        sns.heatmap(confusion_matrix(pre,y_test),annot=True)
        print(model)
        print('accuracy is : ',accuracy,'recall is',recall,'f1 is :', f1)
```

```python
cal(model1)
```

```python
Algorithm=['RandomForestClassifier','DecisionTreeClassifier','GradientBoostingClassifier','KNeighborsClassifier','GaussianNB','XGBRegressor']
R2=[]
RMSE=[]
```

```python
def models(model):
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    r2=r2_score(y_test,pre)
    rmse=np.sqrt(mean_squared_error(y_test,pre))
    R2.append(r2)
    RMSE.append(rmse)
    score=model.score(x_test,y_test)
    print(f"the score of model is :{score}")
```

```python
models(model1)
```

```python
def pred(model):
    model.fit(x_train,y_train)
    pre=model.predict(x_test)
    print(classification_report(pre,y_test))
```
